```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset('iris')
```

```python
df.shape
df.columns
df.head()
df.info()
df.describe()
```

```python
#pairplot 
sns.pairplot(df,hue='species',markers=['o','s','D'],palette='husl',diag_kind='kde', height=2.2, aspect=1 )    
plt.suptitle("Iris Dataset  Pairplot by Species", y=1.02, fontsize=14)
plt.show()
```

```python
df.hist(figsize=(10,8), bins=15, color='pink',edgecolor='blue')
plt.suptitle('histograms of all numeric features', y=1.02)
plt.tight_layout()
plt.show()
```

```python
sns.boxplot(x="species", y="sepal_length", data=df , color='lightgreen')
plt.show()
```
